```python
import xarray as xr
import h3
import h3.unstable.vect
import numpy as np
```

```python
ds = xr.tutorial.load_dataset('air_temperature').load()
ds
```

```python
ds.air[0].plot()
```

```python
resolution = 3

lon, lat = xr.broadcast(ds.lon, ds.lat)
%time index = h3.unstable.vect.geo_to_h3(lat.data.ravel(), lon.data.ravel(), resolution)
index.shape = lon.shape

len(np.unique(index)) / lon.size
```

```python
index.shape
```

```python
ds.lon.shape
```

```python
ds.coords['index'] = ('lat', 'lon'), index.transpose()
ds
```

```python
ds.index.plot()
```

```python
lon_min, lon_max = ds.lon.min().values.item(), ds.lon.max().values.item()
lat_min, lat_max = ds.lat.min().values.item(), ds.lat.max().values.item()

```

```python
import shapely
bbox_coords = [
    (lon_min - 360, lat_min),
    (lon_min - 360, lat_max),
    (lon_max - 360, lat_max),
    (lon_max - 360, lat_min),
    (lon_min - 360, lat_min)
]
bbox = shapely.Polygon(bbox_coords)
bbox
```

```python
bbox_coords
```

```python
# h3 wants lat first
bbox_coords_lat_first = [(lat, lon) for lon, lat in bbox_coords]
bbox_indexes = np.array(list(h3.api.basic_int.polyfill_polygon(bbox_coords_lat_first, resolution)))
bbox_indexes.shape
```

```python
ll_points = np.array([h3.api.basic_int.h3_to_geo(i) for i in bbox_indexes])
ll_points_lon_first = ll_points[:, ::-1]
```

```python
coords={"h3_index": bbox_indexes}

# remember to re-add the 360 degree offset
dsi = ds.interp(
    lon=xr.DataArray(ll_points_lon_first[:, 0] + 360, dims="h3_index", coords=coords),
    lat=xr.DataArray(ll_points_lon_first[:, 1], dims="h3_index", coords=coords)
)
dsi
```

```python
df = dsi.isel(time=0).to_dataframe()
df[df.air.notna()].plot.scatter(x="lon", y="lat", c="air", marker='h', s=20, figsize=(12, 9))
```
